```python
import pandas as pd
import numpy as np
```

```python
workout=pd.read_csv('Workout.csv')
```

```python
workout.head()
```

```python
workout.head()
```

```python
workout['workoutActivityType'].unique()
```

```python
#Renaming workoutActivityType  to Running , Cycling ,Walking , Elliptical ,Jump Rope
workout["workoutActivityType"]=workout["workoutActivityType"].replace("HKWorkoutActivityTypeWalking","Walking")
workout["workoutActivityType"]=workout["workoutActivityType"].replace("HKWorkoutActivityTypeCycling","Cycling")
workout["workoutActivityType"]=workout["workoutActivityType"].replace("HKWorkoutActivityTypeRunning","Running")
workout["workoutActivityType"]=workout["workoutActivityType"].replace("HKWorkoutActivityTypeJumpRope","JumpRope")
workout["workoutActivityType"]=workout["workoutActivityType"].replace("HKWorkoutActivityTypeElliptical","Elliptical")
```

```python
workout["workoutActivityType"].unique()
```

```python
#convert dates to actual datetime
format='%Y-%m-%d %H:%M:%S %z'
```

```python
workout['creationDate'] = pd.to_datetime(workout['creationDate'],format=format)
workout['startDate'] = pd.to_datetime(workout['startDate'], format=format)
workout['endDate'] = pd.to_datetime(workout['endDate'],format=format)
```

```python
#Convert selected columns to numeric
workout[["duration", "totalDistance", "totalEnergyBurned"]] = workout[["duration", "totalDistance", "totalEnergyBurned"]].apply(pd.to_numeric)
```
